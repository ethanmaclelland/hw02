```python
x = np.linspace(-1, 1, 10) #bounds for T(X) (radians)
plt.style.use('fivethirtyeight')
T1 = mycheb(1, x) #n=1
T3 = mycheb(3, x) #n=3
T5 = mycheb(5, x) #n=5

fig, ax = plt.subplots()
ax.plot(x, T1, label=r"$T 1 (x)$", linewidth=3)
ax.plot(x, T3, label=r"$T 3 (x)$", linewidth=3)
ax.plot(x, T5, label=r"$T 5 (x)$", linewidth=3)
ax.legend(loc=2) # upper left corner
fig.set_size_inches(11, 5) #sizing
ax.set_xlabel(r'$x$')
ax.set_ylabel(r'$T(x)$')
ax.set_title('Chebyshev Polynomial at n=1,3,5')
fig.show();
```
